```python
dir_timeseries='F:/data/NASA_biophysical/timeseries_data/'
date_start,date_end='2000-01-01','2018-12-31'
darray = ['oscar','mld','ccmp','sst','sss','ssh']
mask=get_mask()
for itype in range(2,6):
    dtype = darray[itype]
    #read in OSCAR data, put on -180 to 180, -90 to 90 grid, and resample to monthly
    #v6 uses groupby instead of creating climatology first
    ds= get_data(itype,mask)
    ds_tem=ds.sel(time=slice(date_start,date_end))
    data_climatology = ds_tem.groupby('time.month').mean('time',keep_attrs=True,skipna=False)
    data_anomaly = ds_tem.groupby('time.month') - data_climatology
    global_attributes = ds_tem.attrs
    data_anomaly.attrs = global_attributes
    #now iterate over regions
    for i in range(0,9):
        ds_mean = subset_data(ds_tem,i,boxes)
        ds_anom = subset_data(data_anomaly,i,boxes)
        ds_mean_clim = ds_anom + ds_mean

        if i==0:
            ds_box = ds_mean
            ds_box_clim = ds_mean_clim
            ds_box_anom = ds_anom #ds_mean - ds_mean_clim #ds_anom
        else:
            ds_box = xr.concat([ds_box,ds_mean],dim='box')
            ds_box_clim = xr.concat([ds_box_clim,ds_mean_clim],dim='box')
            ds_box_anom = xr.concat([ds_box_anom,ds_anom],dim='box')
    filename_out = dir_timeseries + dtype + 'data_v6.nc'
    ds_box.to_netcdf(filename_out)
    df = ds_box.to_dataframe()
    filename_out = dir_timeseries + dtype + 'data_v6.csv'
    df.to_csv(filename_out)
    filename_out_clim = dir_timeseries + dtype + 'data_minus_clim_v6.nc'
    ds_box_anom.to_netcdf(filename_out_clim)
    df = ds_box_anom.to_dataframe()
    filename_out = dir_timeseries + dtype + 'data_minus_clim_v6.csv'
    df.to_csv(filename_out)
    filename_out_clim = dir_timeseries + dtype + 'data_clim_v6.nc'
    ds_box_clim.to_netcdf(filename_out_clim)
    df = ds_box_clim.to_dataframe()
    filename_out = dir_timeseries + dtype + 'data_clim_v6.csv'
    df.to_csv(filename_out)
```
