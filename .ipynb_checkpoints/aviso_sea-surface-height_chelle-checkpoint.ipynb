# Sea Surface Altimetry Data Analysis

For this example we will use gridded sea-surface altimetry data from The Copernicus Marine Environment:

http://marine.copernicus.eu/services-portfolio/access-to-products/?option=com_csw&view=details&product_id=SEALEVEL_GLO_PHY_L4_REP_OBSERVATIONS_008_047

This is a widely used dataset in physical oceanography and climate.

![globe image](http://marine.copernicus.eu/documents/IMG/SEALEVEL_GLO_SLA_MAP_L4_REP_OBSERVATIONS_008_027.png)

The dataset has already been extracted from copernicus and stored in google cloud storage in [xarray-zarr](http://xarray.pydata.org/en/latest/io.html#zarr) format.

```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import gcsfs
plt.rcParams['figure.figsize'] = (15,10)
%matplotlib inline
```

### Initialize Dataset

Here we load the dataset from the zarr store. Note that this very large dataset initializes nearly instantly, and we can see the full list of variables and coordinates.

```python
import intake
cat = intake.Catalog("https://raw.githubusercontent.com/pangeo-data/pangeo-datastore/master/intake-catalogs/ocean.yaml")
ds = cat["sea_surface_height"].to_dask()
ds
```

### Examine Metadata

For those unfamiliar with this dataset, the variable metadata is very helpful for understanding what the variables actually represent

```python
for v in ds.data_vars:
    print('{:>10}: {}'.format(v, ds[v].attrs['long_name']))
```

## Timeseries of Global Mean Sea Level

Here we make a simple yet fundamental calculation: the rate of increase of global mean sea level over the observational period.

```python
# the number of GB involved in the reduction
ds.sla.nbytes/1e9
```

## Subset data to f&c paper locations at xt and xp

```python
xt_timeseries = ds.sla.sel(latitude=35,longitude=210, method='nearest').load()
xp_timeseries = ds.sla.sel(latitude=55,longitude=210, method='nearest').load()
xt_timeseries_box = ds.sla.sel(latitude=slice(32.5,37.5),longitude=slice(207.5,212.5)).mean(dim=('latitude', 'longitude')).load()
xp_timeseries_box = ds.sla.sel(latitude=slice(52.5,57.5),longitude=slice(207.5,212.5)).mean(dim=('latitude', 'longitude')).load()
```

```python
xt_timeseries.to_netcdf('xt_ts.nc')
xp_timeseries.to_netcdf('xp_ts.nc')
xt_timeseries_box.to_netcdf('xt_ts_box.nc')
xp_timeseries_box.to_netcdf('xp_ts_box.nc')
```

## need to .groupby month & subtract to create anomaly then 3 month boxcar

```python
N=90
climatology = xt_timeseries.groupby('time.month').mean('time')
xt_anomalies = xt_timeseries.groupby('time.month') - climatology
xt_smoothed_anom = np.convolve(xt_anomalies, np.ones((N,))/N, mode='valid')

climatology = xp_timeseries.groupby('time.month').mean('time')
xp_anomalies = xp_timeseries.groupby('time.month') - climatology
xp_smoothed_anom = np.convolve(xp_anomalies, np.ones((N,))/N, mode='valid')
```

```python
xt_timeseries.plot()
xp_timeseries.plot()
```

```python
#show that smoothed timeseries aligned with unsmoothed timeseries
xp_anomalies.plot()
plt.plot(xp_anomalies.time[45:-44],xp_smoothed_anom)
```

```python
plt.plot(xp_anomalies.time[45:-44],xp_smoothed_anom)
plt.plot(xt_anomalies.time[45:-44],xt_smoothed_anom)
#xt_smoothed_anom.plot()
```

```python
plt.plot(xp_anomalies.time[45:-44],xt_smoothed_anom-xp_smoothed_anom)
plt.plot(xp_anomalies.time[45:-44],(xp_smoothed_anom-xt_smoothed_anom)*0)

```

```python
xt_month_ave = xt_timeseries.resample(time='1M').mean('time')
xp_month_ave = xp_timeseries.resample(time='1M').mean('time')
#need to .groupby month & subtract to create anomaly then 3 month boxcar
N=3
climatology = xt_month_ave.groupby('time.month').mean('time')
xt_anomalies = xt_month_ave.groupby('time.month') - climatology
xt_smoothed_anom = np.convolve(xt_anomalies, np.ones((N,))/N, mode='valid')

climatology = xp_month_ave.groupby('time.month').mean('time')
xp_anomalies = xp_month_ave.groupby('time.month') - climatology
xp_smoothed_anom = np.convolve(xp_anomalies, np.ones((N,))/N, mode='valid')
```

```python
plt.plot(xp_anomalies.time[1:-1],xt_smoothed_anom-xp_smoothed_anom)
plt.plot(xp_anomalies.time[1:-1],(xp_smoothed_anom-xt_smoothed_anom)*0)
xmin, xmax = plt.xlim()
plt.xlim(xmin,xmax-4500)
plt.ylabel('$\Delta SSH (m)')
plt.xlabel('Time')
plt.savefig('ssh_2006.png', transparent=False, format='png')
```

```python
plt.plot(xp_anomalies.time[1:-1],xt_smoothed_anom-xp_smoothed_anom)
plt.plot(xp_anomalies.time[1:-1],(xp_smoothed_anom-xt_smoothed_anom)*0)
plt.ylabel('$\Delta SSH (m)')
plt.xlabel('Time')
plt.savefig('ssh_2017.png', transparent=False, format='png')
```

```python
# the computationally intensive step
sla_timeseries = ds.sla.mean(dim=('latitude', 'longitude')).load()
```

```python
sla_timeseries.plot(label='full data')
sla_timeseries.rolling(time=365, center=True).mean().plot(label='rolling annual mean')
plt.ylabel('Sea Level Anomaly [m]')
plt.title('Global Mean Sea Level')
plt.legend()
plt.grid()
```

In order to understand how the sea level rise is distributed in latitude, we can make a sort of [Hovmller diagram](https://en.wikipedia.org/wiki/Hovm%C3%B6ller_diagram).

```python
sla_hov = ds.sla.mean(dim='longitude').load()
```

```python
fig, ax = plt.subplots(figsize=(12, 4))
sla_hov.name = 'Sea Level Anomaly [m]'
sla_hov.transpose().plot(vmax=0.2, ax=ax)
```

We can see that most sea level rise is actually in the Southern Hemisphere.

## Sea Level Variability

We can examine the natural variability in sea level by looking at its standard deviation in time.

```python
sla_std = ds.sla.std(dim='time').load()
sla_std.name = 'Sea Level Variability [m]'
```

```python
ax = sla_std.plot()
_ = plt.title('Sea Level Variability')
```
